# Bubble Sort

```python
from graph import plot_function_performance
from random import sample
```

```python
def bubblesort(elements):
    # Looping from size of array from last index[-1] to index [0]
    for n in range(len(elements)-1, 0, -1):
        swapped = False
        for i in range(n):
            if elements[i] > elements[i + 1]:
                swapped = True
                # swapping data if the element is less than next element in the array
                elements[i], elements[i + 1] = elements[i + 1], elements[i]
        if not swapped:
            # exiting the function if we didn't make a single swap
            # meaning that the array is already sorted.
            return
```

```python
def generate_lists(max_size, step=1):
    """
    Generate a list of lists with sizes increasing linearly.

    Parameters:
    - max_size: The maximum size of the inner lists.
    - step: The step size by which each subsequent list increases in size.

    Returns:
    - A list of lists, where the size of each list increases linearly by the given step size.
    """
    return [sample(range(n), k=len(range(n))) for n in range(1, max_size + 1, step)]
```

```python
inputs = generate_lists(500)

plot_function_performance(func=bubblesort, inputs=inputs, title="Bubble Sort")
```
