# Bubble Sort

```python
from graph import plot_function_performance
from random import sample
```

```python
# Original bubble sort
def bubble_sort_basic(arr: list[int]):
    """
    Basic bubble sort - always does n-1 passes regardless of whether
    the array becomes sorted earlier.
    """
    n = len(arr)

    # Outer loop for number of passes
    for _ in range(n - 1):
        # Inner loop for comparisons in each pass
        for j in range(n - 1):
            # Swap if current element is greater than next
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
def bubble_sort_optimized(arr):
    """
    Optimized bubble sort with two improvements:
    1. Reduces comparison range after each pass (sorted elements bubble to end)
    2. Stops early if no swaps occur (array is already sorted)
    """
    n = len(arr)

    for i in range(n - 1):
        swapped = False

        # Reduce range: last i elements are already sorted
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        # If no swaps occurred, array is sorted
        if not swapped:
            break

    return arr
```

```python
def generate_lists(max_size, step=1):
    """
    Generate a list of lists with sizes increasing linearly.

    Parameters:
    - max_size: The maximum size of the inner lists.
    - step: The step size by which each subsequent list increases in size.

    Returns:
    - A list of lists, where the size of each list increases linearly by the given step size.
    """
    return [sample(range(n), k=len(range(n))) for n in range(1, max_size + 1, step)]
```

```python
inputs = generate_lists(1000)

plot_function_performance(func=bubble_sort_basic, inputs=inputs, title="Bubble Sort")
plot_function_performance(func=bubble_sort_optimized, inputs=inputs, title="Bubble Sort")
```
