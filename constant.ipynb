# Constant Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Constant    | c          | O(c)        | contain only simple statements                         |

```python
from graph import plot_function_performance


def constant(n: float):
    return 2 * n + 1 / 3 - 34


inputs = range(0, 10000000, 100)
plot_function_performance(func=constant, inputs=inputs, title="constant")
```
