# Merge Sort

```python
from graph import plot_function_performance
from random import sample
```

```python
def merge_sort(values):
    if len(values) <= 1:
        return values
    middle = len(values) // 2
    left = values[0:middle]
    right = values[middle:]
    return merge(merge_sort(left), merge_sort(right))


def merge(left, right):
    result = []
    leftIndex = 0
    rightIndex = 0
    while leftIndex < len(left) and rightIndex < len(right):
        if left[leftIndex] < right[rightIndex]:
            result.append(left[leftIndex])
            leftIndex += 1
        else:
            result.append(right[rightIndex])
            rightIndex += 1
    if leftIndex < len(left):
        result += left[leftIndex:]
    elif rightIndex < len(right):
        result += right[rightIndex:]
    return result
```

```python
def generate_random_sized_lists(max_size, step=1):
    """
    Generate a list of lists with sizes increasing linearly.

    Parameters:
    - max_size: The maximum size of the inner lists.
    - step: The step size by which each subsequent list increases in size.

    Returns:
    - A list of lists, where the size of each list increases linearly by the given step size.
    """
    return [sample(range(n), k=len(range(n))) for n in range(1, max_size + 1, step)]
```

```python
inputs = generate_random_sized_lists(1000)

plot_function_performance(func=merge_sort, inputs=inputs, title="Merge Sort")
```
