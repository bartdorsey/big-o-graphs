# Linear Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Linear      | n          | O(n)        | usually contains a single for loop based on input size |

```python
from graph import plot_function_performance


def sum_list(values: list[int]):
    sum = 0
    for value in values:
        sum += value
    return sum


increasing_lists = [list(range(n)) for n in range(0, 100000, 100)]
plot_function_performance(func=sum_list, inputs=increasing_lists, title="sum_list")
```
