# Polynomial Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Polynomial  | n^2        | O(n^2)      | usually has 2 nested loops                             |

```python
from graph import plot_function_performance
```

```python
def create_square_array(size: int):
    result = []
    for i in range(size):
        row = []
        for j in range(size):
            row.append(0)
        result.append(row)
    return result
```

```python
inputs = range(0, 1000)
plot_function_performance(
    func=create_square_array, inputs=inputs, title="create_square_array"
)
```
