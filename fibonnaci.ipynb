# Fibonacci

```python
from graph import plot_function_performance
```

```python
def fibonacci(n: int):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
def fibonacci_iterative(n):
    x = [0,1]
    for i in range(n >> 1):
        x[0] += x[1]
        x[1] += x[0]
    return x[n % 2]
```

```python
inputs = range(1, 35)

plot_function_performance(func=fibonacci, inputs=inputs, title="Fibonacci")
```

```python
inputs = range(1, 1000)

plot_function_performance(func=fibonacci_iterative, inputs=inputs, title="Fibonacci")
```
