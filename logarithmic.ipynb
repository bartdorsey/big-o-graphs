```python
def halving_steps(n: int):
    steps = 0
    while n > 1:
        n = n // 2
        steps += 1
    return steps

inputs = range(0, 100000)

plot_function_performance(func=halving_steps, inputs=inputs, title="halving_steps")

# Example usage
print(halving_steps(8))  # Should print 3 (8 -> 4 -> 2 -> 1)
print(halving_steps(50))  # Should print 5 (50 -> 25 -> 12 -> 6 ->
```

```python
def generate_lists(max_size, step=1):
    """
    Generate a list of lists with sizes increasing linearly.

    Parameters:
    - max_size: The maximum size of the inner lists.
    - step: The step size by which each subsequent list increases in size.

    Returns:
    - A list of lists, where the size of each list increases linearly by the given step size.
    """
    lists = []
    for num in range(1, max_size, step):
        lst = list(range(0, num))
        lists.append((lst, len(lst) // 2))

    return lists 
```

```python

def binary_search(inputs):
    values = inputs[0]
    value_to_find = inputs[1]
    start = 0
    end = len(values) - 1
    while start <= end:
        middle = (start + end) // 2
        if values[middle] == value_to_find:
            return middle
        elif values[middle] < value_to_find:
            start = middle + 1
        else:
            end = middle - 1
    return None


lists = generate_lists(10000)

plot_function_performance(func=binary_search, inputs=lists, title="binary_search")

 
```
