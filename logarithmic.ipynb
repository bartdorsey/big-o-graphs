# Logarithmic Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Logarithmic | log(n)     | O(log n)    | cut things in half over and over                       |

```python
from graph import plot_function_performance
```

```python
def halving_steps(n: int):
    steps = 0
    while n > 1:
        n = n // 2
        steps += 1
    return steps

inputs = range(0, 100000)

plot_function_performance(func=halving_steps, inputs=inputs, title="halving_steps")

# Example usage
print(halving_steps(8))  # Should print 3 (8 -> 4 -> 2 -> 1)
print(halving_steps(50))  # Should print 5 (50 -> 25 -> 12 -> 6 ->
```

```python

def binary_search(values: list[int], value_to_find: int):
    start = 0
    end = len(values) - 1
    while start <= end:
        middle = (start + end) // 2
        if values[middle] == value_to_find:
            return middle
        elif values[middle] < value_to_find:
            start = middle + 1
        else:
            end = middle - 1
    return None



values = list(range(1_234_567)) + [23_842_903_842]
value_to_find = 23_842_903_842
print(binary_search(values, value_to_find))

 
```
