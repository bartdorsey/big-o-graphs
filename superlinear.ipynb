# Superlinear Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Superlinear | n * log(n) | O (n log n) | usually cut things in half and contains a loop         |

```python
from graph import plot_function_performance
```

```python
def count_halves(arr):
    total_steps = 0
    for number in arr:
        steps = 0
        while number > 1:
            number = number // 2
            steps += 1
        total_steps += steps
    return total_steps
```

```python
def generate_linearly_sized_lists(max_size, step=1):
    """
    Generate a list of lists with sizes increasing linearly.

    Parameters:
    - max_size: The maximum size of the inner lists.
    - step: The step size by which each subsequent list increases in size.

    Returns:
    - A list of lists, where the size of each list increases linearly by the given step size.
    """
    return [list(range(n)) for n in range(0, max_size + 1, step)]
```

```python
# Example usage
arr = [30, 40, 50, 60, 70]
generated_lists = generate_linearly_sized_lists(max_size=10000, step=2)

plot_function_performance(
    func=count_halves, inputs=generated_lists, title="Count Halves Performance"
)

total_halving_operations = count_halves(arr)
print(f"Total halving operations: {total_halving_operations}")
```
