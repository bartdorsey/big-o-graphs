# Exponential Time Complexity

| Class       | f(x)       | Big O       | You can find it in things that...                      |
|-------------|------------|-------------|--------------------------------------------------------|
| Exponential | k^n        | O(k^n)      | usually find subsets of a main set                     |

```python
from graph import plot_function_performance

def generate_subsets(input_set):
    subsets = [[]]  # Start with an empty subset

    for element in input_set:
        # For each element, add it to all existing subsets
        new_subsets = []
        for subset in subsets:
            new_subset = subset + [element]
            new_subsets.append(new_subset)
        subsets.extend(new_subsets)

    return subsets



# Example usage:
increasing_lists = [list(range(n)) for n in range(0, 24)]

plot_function_performance(func=generate_subsets, inputs=increasing_lists, title="Generate Subsets")

```
